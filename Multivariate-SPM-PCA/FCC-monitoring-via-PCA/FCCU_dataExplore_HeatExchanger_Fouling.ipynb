## 探索换热器结垢数据集

```python
# 导入包 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 中文显示设置
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False


# 包设置
plt.rcParams.update({'font.size': 14})
```

```python
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#                       获取正常运行工况(NOC)数据
############################################################################### 
```

```python
# 进料变化的正常数据
FCCU_normal_varyingFeed_data = pd.read_csv('NOC_varyingFeedFlow_outputs.csv', header=None)

# 绘制相关变量
plt.figure(figsize=[4,2])
plt.plot(FCCU_normal_varyingFeed_data.iloc[:,1],'seagreen', linewidth=1)
plt.xlabel('时间 (分钟)')
plt.ylabel('F3')
plt.show()

plt.figure(figsize=[4,2])
plt.plot(FCCU_normal_varyingFeed_data.iloc[:,28],'cornflowerblue', linewidth=1)
plt.xlabel('时间 (分钟)')
plt.ylabel('F5')
plt.show()

plt.figure(figsize=[4,2])
plt.plot(FCCU_normal_varyingFeed_data.iloc[:,8],'cornflowerblue', linewidth=1)
plt.xlabel('时间 (分钟)')
plt.ylabel('T3')
plt.show()

plt.figure(figsize=[4,2])
plt.plot(FCCU_normal_varyingFeed_data.iloc[:,22],'cornflowerblue', linewidth=1)
plt.xlabel('时间 (分钟)')
plt.ylabel('V1')
plt.show()
```
