## **主题**: 加热炉数据集探索

```python
# 包
import numpy as np
import matplotlib.pyplot as plt
# 中文显示设置
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

import pandas as pd
```

```python
# 加载无故障数据
faultFree_dataset = pd.read_csv('NoFault_FlowVarying_48HrsData.csv').values
faultFree_dataset_time = faultFree_dataset[:,0]
faultFree_data = faultFree_dataset[:,1:]

faulty_dataset = pd.read_csv('TI1_DisturbanceFault_6HrsData.csv').values
faulty_dataset_time = faulty_dataset[:,0]
faulty_data = faulty_dataset[:,1:]
```

```python
# 两数据集按变量叠加的时间序列图
plt.figure(figsize=(5,8))
plt.subplot(3,1,1)
plt.plot(faultFree_dataset_time, faultFree_data[:,0],'b--',linewidth=3)
plt.plot(faulty_dataset_time, faulty_data[:,0],'r--',linewidth=3)
plt.ylabel('FC1')

plt.subplot(3,1,2)
plt.plot(faultFree_dataset_time, faultFree_data[:,1],'b--',linewidth=3)
plt.plot(faulty_dataset_time, faulty_data[:,1],'r--',linewidth=3)
plt.ylabel('FC2')

plt.subplot(3,1,3)
plt.plot(faultFree_dataset_time, faultFree_data[:,2],'b--',linewidth=3)
plt.plot(faulty_dataset_time, faulty_data[:,2],'r--',linewidth=3)
plt.ylabel('FG')
plt.xlabel('时间 (分钟)')


plt.figure(figsize=(5,12))
plt.subplot(5,1,1)
plt.plot(faultFree_dataset_time, faultFree_data[:,3],'b--',linewidth=3)
plt.plot(faulty_dataset_time, faulty_data[:,3],'r--',linewidth=3)
plt.ylabel('TO')

plt.subplot(5,1,2)
plt.plot(faultFree_dataset_time, faultFree_data[:,4],'b--',linewidth=3)
plt.plot(faulty_dataset_time, faulty_data[:,4],'r--',linewidth=3)
plt.ylabel('FO')

plt.subplot(5,1,3)
plt.plot(faultFree_dataset_time, faultFree_data[:,5],'b--',linewidth=3)
plt.plot(faulty_dataset_time, faulty_data[:,5],'r--',linewidth=3)
plt.ylabel('DT')

plt.subplot(5,1,4)
plt.plot(faultFree_dataset_time, faultFree_data[:,6],'b--',linewidth=3)
plt.plot(faulty_dataset_time, faulty_data[:,6],'r--',linewidth=3)
plt.ylabel('TS1')

plt.subplot(5,1,5)
plt.plot(faultFree_dataset_time, faultFree_data[:,7],'b--',linewidth=3)
plt.plot(faulty_dataset_time, faulty_data[:,7],'r--',linewidth=3)
plt.ylabel('TS2')
plt.xlabel('时间 (分钟)')

plt.show()
```

#### MPC调整FC1以抵消TI1下降对TS1的影响。为将FO维持在所需设定点，MPC增大FC2。
